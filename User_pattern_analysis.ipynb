```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path1 = '/home/gaurav/wns_env/sample-dataset.csv'

features = ['id', 'uuid', 'org_uuid', 'machine', 'observed_at', 'forwarded_at',
       'user', 'application', 'field', 'coord_x', 'coord_y', 'action', 'value',
       'clipboard', 'monetary_amount', 'keystrokes', 'created_at',
       'updated_at', 'screenshot_uuid', 'dom_uuid', 'source']
df1 = pd.read_csv(path1)
df1.head()
```

```python
#preprocessing Data
df = df1.copy()
drop_col = ['coord_y',  'clipboard']
df = df.drop(drop_col,axis = 1)

#Arrange the dataframe in ascending order of observed Time.
df = df.sort_values(by = ['observed_at'],ascending=True)
df['observed_at'] = pd.to_datetime(df['observed_at'])

index = []
for i in range(1,591):
       index.append(i)

#Set the new Index
df['New_Index'] = pd.Series(index)
df.set_index("New_Index",inplace=True)
df.drop(['id'],axis = 1,inplace= True)
df.head()
```

```python
#visualise total time taken in one cycle.
cycle3=  []
cycle4 = []
cycle5 = []
cycle6 = []
for  row in range(1,len(df['cycle_id'])):
       if df.loc[row,'cycle_id'] == 3:
              cycle3.append((row,df.loc[row,'observed_at']))
       elif df.loc[row,'cycle_id'] == 4:
              cycle4.append((row,df.loc[row,'observed_at']))
       elif df.loc[row,'cycle_id'] == 5:
              cycle5.append((row,df.loc[row,'observed_at']))
       elif df.loc[row,'cycle_id'] == 6:
              cycle6.append((row,df.loc[row,'observed_at']))

            
#Divide the cycle Data according to the the length of form fields.
#each cycle contains difeerent kinds of from with different kinds of fields.
#So filling those columns is taking different User-time on each cycle.

list3 = []
from datetime import datetime

for i in range(2,len(cycle3)+1):
       if i % 5 == 0:
              list3.append(cycle3[i-1][0])
list4 = []
for i in range(2,len(cycle4)+1):
       if i % 8 == 0:
              list4.append(cycle4[i-1][0])

list5 = []
for i in range(2,len(cycle5)+1):
       if i % 5 == 0:
              list5.append(cycle5[i-1][0])

list6 = []
for i in range(2,len(cycle3)+1):
       if i % 8 == 0:
              list6.append(cycle6[i-1][0])

```

```python
#calculate the total time spend on each cycle and store them in a list.
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

val = '2017/12/05'
sum = datetime.strptime(val, '%Y/%m/%d')

#Total time taken by cycle3
time_count = []

for i in list3:
       diff = df.loc[i,'observed_at'] - df.loc[i-5,'observed_at']
       sum = sum + diff
time_count.append(sum)


sum = datetime.strptime(val, '%Y/%m/%d')
for i in list4:
       diff = df.loc[i,'observed_at'] - df.loc[i-8,'observed_at']
       sum = sum + diff
time_count.append(sum)


sum = datetime.strptime(val, '%Y/%m/%d')
for i in list5:
       diff = df.loc[i+1,'observed_at'] - df.loc[i-5,'observed_at']
       sum = sum + diff
time_count.append(sum)


sum = datetime.strptime(val, '%Y/%m/%d')
for i in list6:
       diff = df.loc[i,'observed_at'] - df.loc[i-8,'observed_at']
       sum = sum + diff
time_count.append(sum)


x = pd.Series([3,4,5,6])
y = pd.Series([len(cycle3),len(cycle4),len(cycle5),len(cycle6)])
plt.title("Form length in each cycle")
g = sns.barplot(x,y, data=df)

```
